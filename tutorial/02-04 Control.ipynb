# Control

## Conditional Statements

Conditional statements let you control your program's behavior by executing specific blocks of code based on conditions.

### `if` Statements
Use `if`/`else` for simple, binary decisions.

```typescript
const isLoggedIn: boolean = true;
let message: string;
if (isLoggedIn) {
  message = "Welcome back!"; // Stores the welcome message
} else {
  message = "Please log in."; // Stores the prompt to log in
}
message;
```

#### `switch` Statements
Use `switch` for multiple potential outcomes.

```typescript
const userRole: string = "admin";
let accessMessage: string;
switch (userRole) {
  case "admin":
    accessMessage = "You have full access.";
    break;
  case "editor":
    accessMessage = "You can edit content.";
    break;
  case "viewer":
    accessMessage = "You can view content.";
    break;
  default:
    accessMessage = "Unknown role.";
}
accessMessage;
```

## Loops

Loops allow you to repeat a block of code multiple times based on a condition or over a range of values.

### `while` Loops
Use `while` to repeat as long as a condition is `true`.

```typescript
let countdown = 5;
while (countdown > 0) {
  countdown = countdown - 1;
}
countdown;
```

### `for` Loops
Use `for` to iterate over a sequence.

```typescript
let counter: number = 0;
for (let i: number = 1; i <= 5; i++) {
  counter = i;
}
counter;
```

## Loop Control

Control the flow of loops with these keywords:

- `break`: Exits the loop immediately.
- `continue`: Skips the current iteration and moves to the next one.

```typescript
// Break example: Exit loop at a specific point
for (let i: number = 0; i < 10; i++) {
  if (i === 5) {
    break; // Exit the loop when i is 5
  }
}

// Continue example: Skip specific iterations
let i: number = 0
while (true) {
  if (i === 5) {
    break; // Exit the loop when i is 5
  }
  i = i + 1
}
```
